```python

# PERSIAPAN DATA

# download file contoh csv
!wget https://raw.githubusercontent.com/ruang-belajar/python/master/notebook/data/2013-data-kebakaran.csv -O data-kebakaran.csv


import pandas as pd

# load data ke dataframe (df)
df = pd.read_csv("data-kebakaran.csv")
df.head()
```

1. Filter kejadian kebakaran yang terjadi di bulan Maret


```python
maret = df[df['bulan'] == 3]

maret
```

2. Filter kejadian kebakaran yang nilai kerugiannya antara 50jt sampai 100jt


```python
kerugian_50_100 = df[
    (df['kerugian'] >= 50000000) &
    (df['kerugian'] <= 100000000)
    ]


kerugian_50_100
```

3. Tampilkan/Filter 5 kejadian yang tingkat kerugiannya paling besar


```python
kerugian = df.sort_values(by='kerugian', ascending=False)

top_5_kerugian = kerugian.head(5)

top_5_kerugian
```

4. Menggunakan groupby dan count, hitung banyaknya kejadian kebakaran per bulan


```python
kebakaran_per_bulan = df.groupby('bulan')['tanggal'].count()
banyaknya = kebakaran_per_bulan.reset_index(name='jumlah_kejadian')

banyaknya
```

5. Menggunakan groupby dan sum, hitung jumlah kerugian akibat kebakaran per bulan


```python
kerugian_per_bulan = df.groupby('bulan')['kerugian'].sum()
total = kerugian_per_bulan.reset_index(name='total_kerugian')

total
```

6. Menggunakan gruppby dan mean, hitung rata-rata kerugian akibat kebakaran per bulan


```python
rata_kerugian = df.groupby('bulan')['kerugian'].mean()
rata2_kerugian = rata_kerugian.reset_index(name='rata_rata_kerugian')

rata2_kerugian
```

7. Hitung banyaknya kejadian kebakaran yang disebabkan Listrik per bulan

```python
listrik = df[df['penyebab'] == 'Listrik']


kejadian_listrik = listrik.groupby('bulan')['tanggal'].count()
kejadian_listrik_per_bulan = kejadian_listrik.reset_index(name='jumlah_kejadian_listrik')

kejadian_listrik_per_bulan
```

8. Hitung jumlah kerugian akibat kebakaran yang terjadi di kelurahan Kemayoran

```python
kemayoran = df[df['kelurahan'] == 'Kemayoran']


total_kerugian_kemayoran = kemayoran['kerugian'].sum()


print("Jumlah Kerugian di Kelurahan Kemayoran:", total_kerugian_kemayoran)
```

9. Hitung rata-rata kerugian akibat kebakaran yang terjadi di bulan Juni dan Juli,  di Pasar Rebo


```python
pasrebo_juni_juli = df[
    (df['kecamatan'] == 'Pasar Rebo') &
     ((df['bulan'] == 6) | (df['bulan'] == 7))
     ]


rata2_pasrebo_juni_juli = pasrebo_juni_juli['kerugian'].mean()


print("Rata-rata Kerugian di Pasar Rebo pada Bulan Juni dan Juli:", rata2_pasrebo_juni_juli)
```

10. Menggunakan fungsi max, tampilkan nilai kerugian paling besar dari kejadian kebakaran per kelurahan


```python
kerugian_terbesar_per_kelurahan = df.groupby('kelurahan')['kerugian'].max()
kerugian = kerugian_terbesar_per_kelurahan.reset_index(name='nilai_kerugian_terbesar')


kerugian
```
